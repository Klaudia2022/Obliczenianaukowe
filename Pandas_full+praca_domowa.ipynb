````python
pierwsze=pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-abbrevs.csv')
drugie=pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-areas.csv')
trzecie=pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-population.csv')
````

````python
a=drugie.merge(pierwsze,how='outer',on='state')
````

````python
a
````

````python
usa=trzecie.merge(a,how='left', left_on='state/region', right_on='abbreviation')
````

````python
usa
````

````python
usa.isnull().any()
````

kolumna state/region (po zamianie region) i abbrevation mają takie same dane (skróty), więc mogę usunąć kolumnę abbreviation

w kolumnie state zostaną braki danych dla USA, ale USA to nie stan, więc nie zmieniam (tak samo populacja USA i obszar)

Nie ma populacji w Puerto Rico, ale nie jestem w stanie uzupełnić tych danych

````python
usa.rename(columns={'state/region':'region'}, inplace=True)
````

````python
usa.loc[usa.region=='PR', 'state']='Puerto Rico'
````

````python
usa.loc[usa.region=='PR', 'area (sq.mi)']=usa[usa.state=='Puerto Rico']['area (sq. mi)']
````

````python
usa.drop(columns=['abbreviation'], inplace=True)
````

````python
import matplotlib.pyplot as plt

usa2=usa[usa.year==2010]
x=set(usa2['state'])
x=list(x)
x.remove(np.NaN)

under18=[]
over18=[]
for i in range(len(x)):
  under18.append(float(usa2[(usa2['state']==x[i])&(usa2['ages']=='under18')]['population']))
  over18.append(float(usa2[(usa2['state']==x[i])&(usa2['ages']=='total')]['population'])-under18[i])

plt.figure(figsize=(15, 7))
plt.bar(x, under18, color = 'red', alpha=0.5, label = 'under18')
plt.bar(x, over18, color = 'green', alpha=0.5, label = 'over18')
 
plt.xticks(x,x, rotation=90)
 
plt.legend()
plt.xlabel('States')
plt.ylabel('Liczba ludności w 10')
````

````python
gestosc={}
for el in x:
  gestosc[el]=float(usa2[(usa2['state']==el)&(usa2['ages']=='total')]['population'])/float(usa2[(usa2['state']==el)&(usa2['ages']=='total')]['area (sq. mi)'])

a=sorted(gestosc.items(), key=lambda x: x[1], reverse=True)
state=[]
gestosc2=[]

for el in a:
  state.append(el[0])
  gestosc2.append(el[1])

y_pos = np.arange(len(gestosc2))

plt.figure(figsize=(15,7))
plt.bar(y_pos, gestosc2, color = 'red', alpha=0.5)
plt.xticks(y_pos, state, rotation=90)
plt.xlabel('State')
plt.ylabel('Gęstość')
````
