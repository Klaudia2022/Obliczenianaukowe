## Apply, groupby, agg

```python
def kategoryzuj_populacje(ludnosc):
  if ludnosc <10000:
    return 'małe miasto'
  elif ludnosc <100000:
    return 'średnie miasto'
  else:
    return 'duże miasto'
  
```

```python
dane['Wielkośc miasta']=dane.population.apply(kategoryzuj_populacje)
dane.head()
```

```python
def polozenie(x,y):
  return f'({x},{y})'

dane[['lat','lng']]
```

```python
dane[['lat','lng']].apply(polozenie)
```

```python
# operacje na wierszu
def polozenie(wiersz):
  wiersz['polozenie']=f'({wiersz.lat},{wiersz.lng})'
  return wiersz
dane.apply(polozenie,axis=1)
```

```python
dane.drop(columns=['Apparent Temperature (C)','Wind Bearing (degrees)','Daily Summary'],inplace=True)
```

```python
def wytnij(fdata,a,b,convert=False):
  if convert:
    return int(fdata[a:b])
  else:
    return fdata[a:b]
dane['data']=dane['Formatted Date'].apply(wytnij,args=(0,10))
dane['rok']=dane['data'].apply(wytnij,args=(0,4,True))
dane['miesiac']=dane['data'].apply(wytnij,args=(5,7,True))

def wytnij_date(fdata):
  return fdata[:10]

# inna wersja
dane['new_date']=pd.to_datetime(dane['Formatted Date'])
dane['year']=dane['new_date'].apply(lambda x:x.year)
```

```python
dane.head()
```

```python
Ddane=dane.groupby('data')
dane_dzienne_avg=Ddane.mean()
dane_dzienne_avg.to_excel('dane_dzienne.xlsx')
dane_dzienne_avg
```

```python
Maj_daily=dane[(dane['rok']==2010) & (dane['miesiac']==5) ].groupby('data')
Maj_minimalne=Maj_daily.min()['Temperature (C)']
Maj_maksymalne=Maj_daily.max()['Temperature (C)']
```
