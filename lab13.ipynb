```python
import skimage
from skimage import data, io, filters
import numpy as np
import matplotlib.pyplot as plt
plt.set_cmap('gray')
```

```python
image=data.camera()
print(image.shape)
```

```python
image2=filters.gaussian(image)
```

```python
image3=filters.gaussian(image, sigma=3)
```

```python
fig,ax=plt.subplots(1,3, figsize=(20,8))
ax[0].imshow(image)
ax[1].imshow(image2)
ax[2].imshow(image3)
```

```python
image22=filters.gaussian(image2)
r=image2-image22
ostry=image2+4*r
```

```python
fig,ax=plt.subplots(1,1,figsize=(5,5))
ax.imshow(ostry)
```

```python
tekst=data.text()
```

```python
tekst1=filters.sobel(tekst)
```

```python
tekst2=filters.sobel_h(tekst)
```

```python
tekst3=filters.sobel_v(tekst)
```

```python
tekst4=filters.roberts(tekst)
```

```python
fig,ax=plt.subplots(1,5,figsize=(30,10))
ax[0].imshow(tekst)
ax[1].imshow(tekst1)
ax[2].imshow(tekst2)
ax[3].imshow(tekst3)
ax[4].imshow(tekst4)
```
