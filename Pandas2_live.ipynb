## Apply, groupby, agg

```python
def kategoryzuj_populacje(ludnosc):
  if ludnosc<10000:
    return 'male miasto'
  elif ludnosc<100000:
    return 'srednie miasto'
  else:
    return 'duze miasto'
```

```python
dane['wielkosc miasta']=dane.population.apply(kategoryzuj_populacje)
```
